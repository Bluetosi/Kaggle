```python
import numpy as np
import pandas as pd
import missingno as msno

import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

import sklearn.preprocessing

from sklearn.preprocessing import scale, robust_scale, minmax_scale, normalize

data = pd.read_csv('./Input/train.csv')
```

```python
data.describe()
```

```python
# ID 처리
data_id = data['id']
data.drop('id', axis=1, inplace=True)
data.head(5)
```

```python
from scipy.stats import spearmanr

plt.figure(figsize=(21,21))
sns.set(font_scale=1.25)
sns.heatmap(data.corr(method='spearman'),fmt='.2f', annot=True, square=True , annot_kws={'size' : 15})
```

```python
cor = data.corr(method='spearman')
cor["price"].nlargest(n=20).index
# grade, sqft_living, sqft_above, bathrooms, lat, bedrooms, floors, view, waterfront, condition
```

```python
# 날짜 변경 (년.월.일)
data['date'] = data['date'].apply(lambda x: str(x[0:8])).astype(int)
data.head(5)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
g = sns.distplot(data['price'], color='b', label='Skewness : {:.2f}'.format(data['price'].skew()), ax=ax)
g = g.legend(loc='best')
```

```python
fig = plt.figure(figsize = (15,8))

fig.add_subplot(1,3,1)
res = stats.probplot(data['price'], plot=plt)

fig.add_subplot(1,3,2)
res = stats.probplot(np.log(data['price']), plot=plt)

fig.add_subplot(1,3,3)
```
